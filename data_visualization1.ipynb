```python
alt.Chart(otto_storm).mark_point(filled=True, size=100).encode(
    alt.X('wind:Q', axis=alt.Axis(title= "Storm's maximum sustained wind speed (in knots)"), scale=alt.Scale(zero=False, domain=(25, 110))),
    alt.Y('pressure:Q', axis=alt.Axis(title="Air pressure at the storm's center (in millibars)"), scale=alt.Scale(zero=False)),
    alt.Color('status:N'),
    alt.Shape('category:N'),
    tooltip = [
        alt.Tooltip("year:N"),
        alt.Tooltip('month:N'),
        alt.Tooltip('day:N'),
        alt.Tooltip('hour:N')]
).properties(title='Otto Storm: Wind and Pressure Data')
```

I am showing the storm’s maximum sustained wind speed (in knots), its air pressure at the storm’s center (in millibars), its storm classification status, and Saffir-Simpson storm category. I encoded the wind speed on the x-axis quantitative value; the air pressure on the y-axis quantitative value; the status with the color; and the category with the shape of the scatter plot points.

## D.

The wind and the pressure seem to be in an inverse relationship. As the storm's maximum sustained wind speed increases, the air pressure at the storm’s center decreases. The storm's status category seem to have an impact on the range of these values, and not on the inverse relationship. It seems that at low wind speed the category is tropical depression, and as the wind spees increases to 35-60 interval, the category is tropical storm. At high speeds (60+ knots), the category is hurricane. Category also seems to have an impact on the range. However, this is not surprising, as in the description of the dataset (source="https://vincentarelbundock.github.io/Rdatasets/doc/dplyr/storms.html"), it is stated that the category is estimated from wind speed.

# PROBLEM 2

```python
df = pd.read_csv("Current_Employee_Names__Salaries__and_Position_Titles.csv")
```

```python
df.head(10)
```

```python
df["Salary or Hourly"][1]
```

```python
weekly = []
for i in range(len(df)):
    if df["Salary or Hourly"][i] == "Salary":
        weekly.append(df["Annual Salary"][i] / 52)
    else:
        weekly.append(df["Typical Hours"][i] * df["Hourly Rate"][i])
```

```python
df["Weekly Pay"] = weekly
```

```python
df
```

```python
new = df.groupby(by="Department").sum().reset_index()
new.sort_values(by="Weekly Pay", ascending=False)
```

```python
value_counts = df.Department.value_counts().reset_index()
value_counts['Counts'] = value_counts['Department']
value_counts = value_counts.drop(["Department"], axis=1)
value_counts = value_counts.rename(columns={"index": "Department"})
value_counts
```

```python
merged_df = new.merge(value_counts, how = 'left', on = ['Department'])
merged_df
```

```python
alt.data_transformers.disable_max_rows()

alt.Chart(merged_df).mark_bar().encode(
    alt.X('Department:N', sort='-y', axis=alt.Axis(title= "Employee Department")),
    alt.Y('Weekly Pay:Q', axis=alt.Axis(format='$', title="Total Weekly Pay of the Department")),
    #alt.Size('Counts:Q', scale=alt.Scale(range=[0, 12000])),
    alt.Tooltip('Counts:Q')
).properties(title='City of Chicago Employees Departmental Weekly Pay Distribution', width=700, height=500).configure_axis(titleFontSize=15)
```

Figure I. In this visualization, the City of Chicago's current employees' Departmental Total Weekly Pays are presented with the specific department of the employees presented on the x-axis. When hovered over the bars, we can also see the counts of the employees in the department. It can be seen from the graph that, most of the weekly budget is spent on the departments with more employees.

My question was "How is the budget distribution of the City of Chicago within its departments?". However, it was impossible for me to find this directly, as there are two payment types: hourly and annual. The City of Chicago's website states that (source: "https://data.cityofchicago.org/Administration-Finance/Current-Employee-Names-Salaries-and-Position-Title/aned-ke5c"), "annual wages will vary by employee, depending on number of hours worked and seasonal status." So, I couldn't calculate the annual budget distribution with the given information. Instead, for hourly workers, I calcualted their weekly salary by multiplying their weekly working hours and their hourly payments. For people with an annual salary, I calcualted their weekly pays by dividing their annual salary to 52. This way, with the given information, I calculated the minimum payment unit available for all employees. Then, I grouped the values by each unique department, summed the department workers' weekly wages, and counted the number of employees in each department. Then, I graphed the Departmental Total Weekly Pays and Department, with an interactive count widget. By looking at the visual, we can see that most of the weekly budget is spent for police, fire, water management, streets, aviation, and transportation departments, which is not surprising, as these departments have the highest number of employees. It can be seen from the graph by hovering over the bars that, most of the weekly budget is spent on the departments with more employees.
